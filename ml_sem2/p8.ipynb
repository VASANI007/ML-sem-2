```python
# Attempt following questions based on Pandas
# a. Create a Pandas DataFrame with columns Roll_No, Name, and Marks for 5 students.
# b. Given a DataFrame df, write a statement to display the first 3 rows.
# c. From a DataFrame df, display only the Marks column
# d. Add a new column Result to a DataFrame where Result is "Pass" if Marks >= 40, else "Fail".
# e. Write a Pandas command to check for missing values in a DataFrame.
import pandas as pd
# a. Create a Pandas DataFrame for 5 students
data = {
"Roll_No": [1, 2, 3, 4, 5],
"Name": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
"Marks": [78, 35, 56, 42, 29]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print()
# b. Display the first 3 rows of the DataFrame
print("First 3 rows of DataFrame:")
print(df.head(3))
print()
# c. Display only the Marks column
print("Marks column:")
print(df["Marks"])
print()
# d. Add a new column Result (Pass if Marks >= 40 else Fail)
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")
print("DataFrame after adding Result column:")
print(df)
print()
# e. Check for missing values in the DataFrame
print("Missing values in DataFrame:")
print(df.isnull())
```
