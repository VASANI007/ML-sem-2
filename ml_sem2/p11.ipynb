```python
# Create a CSV file named employees_p11.csv containing the following columns:
# Emp_ID, Name, Department, Salary. Write a Python program using the
# Pandas library to perform the following tasks:
# a. Read the data from the employees.csv file into a Pandas DataFrame.
# b. Display all employee records belonging to the IT department.
# c. Display all employee records whose Salary is not equal to 40,000.
import pandas as pd
# a. Read data from employees_p11.csv into a DataFrame
df = pd.read_csv("employees_p11.csv")
print("Complete Employee Data:")
print(df)
print()
# b. Display all employee records belonging to the IT department
print("Employees from IT Department:")
print(df[df["Department"] == "IT"])
print()
# c. Display all employee records whose Salary is not equal to 40000
print("Employees with Salary not equal to 40000:")
print(df[df["Salary"] != 40000])
```
