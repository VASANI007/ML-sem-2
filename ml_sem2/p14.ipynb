```python
# Handling Missing Values Using Mean- A DataFrame contains a column Marks with some
# missing values.Replace all missing values with the mean of the Marks column and
# display the updated DataFrame.
import pandas as pd
import numpy as np
# Sample DataFrame with missing values
data = {
"Roll_No": [1, 2, 3, 4, 5],
"Name": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
"Marks": [78, np.nan, 56, np.nan, 64]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print()
# Calculate mean of Marks column
mean_marks = df["Marks"].mean()
print("Mean of Marks:", mean_marks)
print()
# Replace missing values with mean (Correct method)
df["Marks"] = df["Marks"].fillna(mean_marks)
print("Updated DataFrame after replacing missing values:")
print(df)
```
