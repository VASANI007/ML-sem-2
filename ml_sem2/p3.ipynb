```python
# Perform following operations based on NumPy – Array Creation & Properties
# a. Create a NumPy array using arange() from 0 to 20 with step 2
# b. Find the data type of a NumPy array
# c. Change the data type of an array from int to float
# d. Find the length of a NumPy array
import numpy as np
# a. Create a NumPy array using arange() from 0 to 20 with step 2
arr = np.arange(0, 21, 2)
print("Array from 0 to 20 with step 2:", arr)
# b. Find the data type of the NumPy array
print("Data type of array:", arr.dtype)
# c. Change the data type of array from int to float
arr_float = arr.astype(float)
print("Array after changing data type to float:", arr_float)
print("New data type:", arr_float.dtype)
# d. Find the length of the NumPy array
print("Length of the array:", len(arr))
```
