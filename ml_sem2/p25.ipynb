```python
# Write a Python program using Scikit-learn to classify students as Pass or Fail based on their marks using the K-Nearest Neighbors (KNN) algorithm. 
import numpy as np 
import pandas as pd 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score, classification_report 
# Step 1: Create dataset 
data = { 
"Marks": [25, 35, 45, 55, 65, 75, 85, 30, 50, 70], 
"Result": ["Fail", "Fail", "Pass", "Pass", "Pass", "Pass", "Pass", "Fail", 
"Pass", "Pass"] 
} 
df = pd.DataFrame(data)
# Convert Result to numeric 
df["Result"] = df["Result"].map({"Fail": 0, "Pass": 1})
```

```python
# Step 2: Features and labels 
X = df[["Marks"]] 
y = df["Result"] 
# Step 3: Train-test split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 4: Train KNN model 
knn = KNeighborsClassifier(n_neighbors=3) 
knn.fit(X_train, y_train) 
# Step 5: Predict test data 
y_pred = knn.predict(X_test) 
# Step 6: Evaluate model 
print("Accuracy:", accuracy_score(y_test, y_pred)) 
print("\nClassification Report:") 
print(classification_report(y_test, y_pred))
```

```python
# Step 7: Predict new student result (Correct method) 
new_marks = pd.DataFrame([[48]], columns=["Marks"]) 
prediction = knn.predict(new_marks) 
if prediction[0] == 1: 
    print("Student Result: Pass") 
else: 
    print("Student Result: Fail")
```
