```python
# Draw a line graph to compare temperatures of two cities over a week.
import matplotlib.pyplot as plt
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
city1_temp = [32, 34, 33, 35, 36, 37, 38]
city2_temp = [28, 30, 29, 31, 32, 33, 34]
plt.plot(days,city1_temp, label="city1_temp", marker = "o")
plt.plot(days,city2_temp, label="city1_temp", marker = "*")
plt.title("Weekly Temperature Comparison of Two Cities")
plt.xlabel("Days")
plt.ylabel("City Temperature")
plt.legend()
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.show()
```
