```python
# Change the value of K from 3 to 5 and observe the change in classification result. 
import numpy as np 
from sklearn.neighbors import KNeighborsClassifier 
# Step 1: Simple dataset 
# Feature: Marks 
X = np.array([[35], [45], [55], [65], [75], [85]]) 
# Labels: 0 = Fail, 1 = Pass 
y = np.array([0, 0, 1, 1, 1, 1]) 
# New data point to classify 
new_data = np.array([[50]])
```

```python
# K = 3 
knn_3 = KNeighborsClassifier(n_neighbors=3) 
knn_3.fit(X, y) 
result_k3 = knn_3.predict(new_data) 
# K = 5 
knn_5 = KNeighborsClassifier(n_neighbors=5) 
knn_5.fit(X, y) 
result_k5 = knn_5.predict(new_data)
```

```python
# Display results 
print("Result with K = 3:", "Pass" if result_k3[0] == 1 else "Fail") 
print("Result with K = 5:", "Pass" if result_k5[0] == 1 else "Fail") 
```
