```python
# Perform following operations based on NumPy – Mathematical Operations
# a. Find the square of each element in a NumPy array
# b. Find the maximum element row-wise in a 2D array
# c. Calculate the standard deviation of an array
# d. Find the index of the maximum element in an array
import numpy as np
# Create a sample NumPy array
arr = np.array([2, 4, 6, 8, 10])
print("Original Array:", arr)
# a. Find the square of each element in the array
square_arr = np.square(arr)
print("Square of each element:", square_arr)
print()
# Create a 2D array for row-wise operations
arr_2d = np.array([
[10, 20, 30],
[5, 15, 25],
[7, 14, 21]
])
print("2D Array:")
print(arr_2d)
# b. Find the maximum element row-wise in a 2D array
row_max = np.max(arr_2d, axis=1)
print("Row-wise maximum elements:", row_max)
print()
# c. Calculate the standard deviation of an array
std_dev = np.std(arr)
print("Standard Deviation of array:", std_dev)
print()
# d. Find the index of the maximum element in an array
max_index = np.argmax(arr)
print("Index of maximum element:", max_index)
```
