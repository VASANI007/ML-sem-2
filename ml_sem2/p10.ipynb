```python
# Create a CSV file named students.csv containing the following columns:
# Roll_No, Name, Marks, Result. Write a Python program using the
# Pandas library to perform the following operations on the DataFrame:
# a. Load the data from the students.csv file into a Pandas DataFrame.
# b. Display all student records where the Marks obtained are greater than 60.
# c. Display the records of students whose Marks are between 50 and 80.
# d. Display all student records where the Result is "Fail".
import pandas as pd
# a. Load data from students.csv into a DataFrame
df = pd.read_csv("students_p10.csv")
print("Complete DataFrame:")
print(df)
print()
# b. Display records where Marks > 60
print("Students with Marks greater than 60:")
print(df[df["Marks"] > 60])
print()
# c. Display records where Marks are between 50 and 80
print("Students with Marks between 50 and 80:")
print(df[(df["Marks"] >= 50) & (df["Marks"] <= 80)])
print()
# d. Display records where Result is 'Fail'
print("Students with Result = Fail:")
print(df[df["Result"] == "Fail"])
```
