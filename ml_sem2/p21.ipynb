```python
# Plot two different graphs in the same figure. 
import matplotlib.pyplot as plt 
x = [1, 2, 3, 4, 5] 
bar_values = [5, 15, 25, 35, 45] 
y_bar = [5, 15, 25, 35, 45] 
# Create a figure with 2 subplots 
```

```python
# Create figure
plt.figure(figsize=(8, 6))

# Line graph (subplot 1)
plt.subplot(1, 2, 1)
plt.bar(x, y_bar)
plt.title("Bar Graph")
plt.xlabel("X Values")
plt.ylabel("Line Values")

# Pie chart (subplot 2)
plt.subplot(1, 2, 2)
plt.pie(bar_values, labels=x, autopct='%1.1f%%')  
plt.title("Pie Chart")

# Adjust layout and display
plt.tight_layout()
plt.show()
```
