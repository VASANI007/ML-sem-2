```python
# Attempt following questions based on matplotlib library  
# a. Write a program to plot a line graph showing the relationship between  two lists of values.  
# b. Write a program to draw a line graph and add: Title, X-axis label, Y-axis label  
# c. Write a Program to plot two or more lines on the same graph with labels. 
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y1 = [9,36,35,52,25]
y2 = [5,22,19,46,55]
plt.plot(x,y1)
plt.show()
```

```python
plt.plot(x,y2, marker = "o")
plt.title("Student Marrks Graph")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
plt.plot(x,y1, label="Line 1", marker = "o")
plt.plot(x,y2, label="Line 2", marker = "*")
plt.title("MultiLine Graph")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.show()
```
