```python
# Given a DataFrame having Department and Salary columns, write a Pandas program to calculate:
# a. Total salary
# b. Average salary for each department.
import pandas as pd
# Sample DataFrame
data = {
"Department": ["IT", "HR", "IT", "Finance", "HR", "Finance"],
"Salary": [50000, 40000, 60000, 55000, 45000, 58000]
}
df = pd.DataFrame(data)
print("Employee Salary Data:")
print(df)
print()
# a. Calculate total salary
total_salary = df["Salary"].sum()
print("Total Salary:", total_salary)
print()
# b. Calculate average salary for each department
avg_salary_dept = df.groupby("Department")["Salary"].mean()
print("Average Salary for each Department:")
print(avg_salary_dept)
```
