```python
# Remove Duplicate Records. A DataFrame contains duplicate student records.
#Write Pandas program to identify and remove duplicate rows.
import pandas as pd
# Sample DataFrame with duplicate records
data = {
"Roll_No": [1, 2, 3, 2, 4, 1],
"Name": ["Amit", "Neha", "Rahul", "Neha", "Priya", "Amit"],
"Marks": [78, 65, 82, 65, 74, 78]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print()
# Identify duplicate rows
print("Duplicate rows (True means duplicate):")
print(df.duplicated())
print()
# Display only duplicate records
print("Duplicate Records:")
print(df[df.duplicated()])
print()
# Remove duplicate rows
df_cleaned = df.drop_duplicates()
print("DataFrame after removing duplicate records:")
print(df_cleaned)
```
