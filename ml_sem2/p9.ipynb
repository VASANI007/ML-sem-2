```python
# Attempt following questions based on Pandas
# a. From a DataFrame containing student marks, select all students who scored more than 70 marks.
# b. Sort the DataFrame in descending order of Marks.
# c. Given a DataFrame with columns Department and Salary, find the average salary of each department.
# d. Replace missing values in the Marks column with the average marks of the class.
# e. Write Pandas code to read a CSV file named students.csv and display its basic information (rows, columns, data types).
import pandas as pd
import numpy as np
# Create a sample DataFrame of students
data = {
"Roll_No": [1, 2, 3, 4, 5, 6],
"Name": ["Amit", "Neha", "Rahul", "Priya", "Karan", "Sneha"],
"Marks": [78, 65, 82, np.nan, 55, 91]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print()
# a. Students scoring more than 70 marks
print("Students scoring more than 70 marks:")
print(df[df["Marks"] > 70])
print()
# b. Sort by Marks descending
print("Sorted DataFrame:")
print(df.sort_values(by="Marks", ascending=False))
print()
# c. Average salary by department
salary_data = {
"Department": ["IT", "HR", "IT", "Finance", "HR", "Finance"],
"Salary": [50000, 40000, 60000, 55000, 45000, 58000]
}
df_salary = pd.DataFrame(salary_data)
print("Average Salary by Department:")
print(df_salary.groupby("Department")["Salary"].mean())
print()
# d. Replace missing Marks with mean
avg_marks = df["Marks"].mean()
df["Marks"] = df["Marks"].fillna(avg_marks)
print("Updated DataFrame:")
print(df)
print()
# e. Read CSV and display info
df_csv = pd.read_csv("students_p9.csv")
print("CSV File Information:")
df_csv.info()
```
