```python
# write a Python program to cluster students based on their marks using the K-Means clustering algorithm. 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
# Step 1: Create sample dataset 
data = { 
"Marks": [35, 40, 45, 55, 60, 65, 70, 80, 85, 90] 
} 
df = pd.DataFrame(data) 
print("Student Marks Data:") 
print(df) 
print() 
```

```python
# Step 2: Apply K-Means Clustering 
# Number of clusters (e.g., Low, Medium, High) 
kmeans = KMeans(n_clusters=3, random_state=42) 
df["Cluster"] = kmeans.fit_predict(df[["Marks"]]) 
print("Data after K-Means Clustering:") 
print(df)  
print() 
```

```python
# Step 3: Visualize the clusters 
plt.scatter(df["Marks"], df["Cluster"]) 
plt.title("K-Means Clustering of Students Based on Marks") 
plt.xlabel("Marks") 
plt.ylabel("Cluster") 
plt.show()
```
