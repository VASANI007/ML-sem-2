```python
# A DataFrame df contains employee details with columns Emp_ID,Name, Department, and
# Salary.Write a Pandas program to display all employees whose salary is between 30,000
# and 60,000.
import pandas as pd
# Sample employee DataFrame
data = {
"Emp_ID": [101, 102, 103, 104, 105, 106],
"Name": ["Amit", "Neha", "Rahul", "Priya", "Karan", "Sneha"],
"Department": ["IT", "HR", "IT", "Finance", "IT", "HR"],
"Salary": [25000, 40000, 55000, 60000, 45000, 70000]
}
df = pd.DataFrame(data)
print("Employee Data:")
print(df)
print()
# Display employees with salary between 30,000 and 60,000
print("Employees with Salary between 30,000 and 60,000:")
print(df[(df["Salary"] >= 30000) & (df["Salary"] <= 60000)])
```
